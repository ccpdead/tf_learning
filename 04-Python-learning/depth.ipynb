```python
import cv2
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
image_path = "/home/jhr/depth_image1.png"
csv_path="/home/jhr/depth.csv"
```

```python
img_input=cv2.imread(image_path,cv2.IMREAD_GRAYSCALE)
mask = np.zeros_like(img_input,dtype=np.uint8)
data = pd.read_csv(csv_path)
data=data.astype(int)
```

```python
for i in range(len(data)):
    cv2.rectangle(img_input, (data.x[i], data.y[i]), (data.x2[i], data.y2[i]), (0, 255, 0), 1)
    radius=max(int((-data.x[i]+data.x2[i])/2),int((-data.y[i]+data.y2[i])/2))
    
    # cv2.circle(mask,
    #        (int((data.x[i]+data.x2[i])/2),int((data.y[i]+data.y2[i])/2)),
    #        int(radius/6)*6,
    #        (50,50,50),
    #        -1)
    
    cv2.circle(mask,
               (int((data.x[i]+data.x2[i])/2),int((data.y[i]+data.y2[i])/2)),
               int(radius/6)*3,
               (255,255,255),
               -1)
    mask_image = cv2.bitwise_and(img_input,mask)
    mask_image2=cv2.bitwi
    mean_color=np.mean(mask_image,axis=(0,1))
    
    cv2.imshow("mean color",mask_image)
    cv2.imshow("mask",mask)
    
    print("mean color:{}".format(mean_color))
    
    # for r in range(1,7,1):
    #     cv2.circle(img_input,
    #                (int((data.x[i]+data.x2[i])/2),int((data.y[i]+data.y2[i])/2)),
    #                (int(radius/6)*r),
    #                (0,255,0),
    #                1)
```
